```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split

data_bank = pd.read_csv('../data/UniversalBank.csv')
# data_bank = pd.read_csv('../code3/data/UniversalBank.csv') # Krima
# data_bank = pd.read_csv('/Users/arturoperezespinosa/Downloads/UniversalBank.csv') # Arthur

data_bank.head(5)
```

```python
data_bank = data_bank.drop(['ID', 'ZIP Code'], axis=1)

data_bank.head(5)
```

```python
data_bank.describe()
```

```python
dummies = pd.get_dummies(data_bank['Education'])

dummies.columns = ['Education_1', 'Education_2', 'Education_3']

data_bank = pd.concat([data_bank, dummies], axis=1)

data_bank = data_bank.drop(['Education'], axis=1)

data_bank.rename(columns={'1':'Education_1', '2': 'Education_2', '3': 'Education_3'})

data_bank.head(5)
```

```python
X_train, X_val, = train_test_split(data_bank, test_size=0.4, random_state= 1)

print(X_train.shape, X_val.shape)
```

```python
y_train = data_bank.loc[X_train.index, "Personal Loan"]

X_train = X_train.drop("Personal Loan", axis=1)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
```

```python
new_customer = {
    'Age': 40,
    'Experience': 10,
    'Income': 84,
    'Family': 2,   
    'CCAvg': 2,
    'Mortgage': 0,
    'Securities Account': 0,
    'CD Account': 0,
    'Online': 1,
    'CreditCard': 1,
    'Education_1': 0,
    'Education_2': 1,
    'Education_3': 0
}

df_nc = pd.DataFrame([new_customer])

df_ncs = scaler.transform(df_nc)

knn = KNN(n_neighbors=1)

knn.fit(X_train_scaled, y_train)

prediction = knn.predict(df_ncs)

if prediction[0] == 1:

    print("Prediction: Customer will get a loan.")

else:
    
    print("Prediction: Customer will NOT get a loan.")
```

Q 1.b

```python
best_k = None

best_test_acc = 0

best_train_test_gap = float("inf")

y_val = X_val["Personal Loan"]

X_val = X_val.drop("Personal Loan", axis=1)

X_val_scaled = scaler.transform(X_val)

K_val = [1, 3, 9, 36]

K_val_filtered = [k for k in K_val if k > 1 and k < 36]

for K in K_val:
    
    knn = KNN(n_neighbors=K)

    knn.fit(X_train_scaled, y_train)

    train_acc = knn.score(X_train_scaled, y_train)

    test_acc = knn.score(X_val_scaled, y_val)

    gap = train_acc - test_acc  

    print(f'Training accuracy of K={K} is {train_acc:.3f}')

    print(f'Test accuracy of K={K} is {test_acc:.3f}')
    
    if K in K_val_filtered:

      if test_acc > best_test_acc:  
            
            best_k = K
            
            best_test_acc = test_acc
            
            best_gap = gap  
      
      elif test_acc == best_test_acc and gap < best_train_test_gap:  
            
            best_k = K

            best_train_test_gap = gap  

print(f'\nBest K is: {best_k} with test accuracy={best_test_acc:.3f} and test-train gap={best_gap:.3f}')
    
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

data_bank_lda = pd.read_csv('../data/UniversalBank.csv')
data_bank_lda = data_bank_lda.drop(['ID','ZIP Code'], axis=1)

dummies = pd.get_dummies(data_bank_lda['Education'])
dummies.columns = ['Education_1', 'Education_2', 'Education_3']
data_bank_lda = pd.concat([data_bank_lda, dummies], axis=1)
data_bank_lda = data_bank_lda.drop(['Education'], axis=1)
data_bank_lda.rename(columns={'1':'Education_1', '2': 'Education_2', '3': 'Education_3'})
```

```python
X = data_bank_lda.drop('Personal Loan', axis=1)
y = data_bank_lda['Personal Loan']

X_train, X_val, y_train, y_val = train_test_split(X, y, 
                                                  test_size=0.4, 
                                                  random_state=100, 
                                                  stratify=y)


continuous_vars = ['Age', 'Experience', 'Income', 'CCAvg', 'Mortgage']

train_data = X_train.copy()
train_data['Personal Loan'] = y_train

summary_continuous = train_data.groupby('Personal Loan')[continuous_vars].agg(['mean', 'std'])
print("\nSummary Statistics for Continuous Predictors:")
print(summary_continuous)

categorical_vars = ['Family', 'Securities Account', 'CD Account', 'Online', 'CreditCard', 'Education_1', 'Education_2', 'Education_3'] 

print("\nSummary Statistics for Categorical Predictors (Percentage Distribution):")
for var in categorical_vars:
    print(f"\nVariable: {var}")
    
    percentages = train_data.groupby('Personal Loan')[var].value_counts(normalize=True).rename('Percentage').mul(100)
    print(percentages)
```

```python
lda = LDA()
lda.fit(X_train, y_train)

y_pred = lda.predict(X_val)
y_pred_proba = lda.predict_proba(X_val)[:, 1]  


accuracy = accuracy_score(y_val, y_pred)
print("\nModel Performance on Validation Set:")
print("Accuracy: {:.2f}%".format(accuracy * 100))

conf_matrix = confusion_matrix(y_val, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
tn, fp, fn, tp = conf_matrix.ravel()
print("\nMisclassification Analysis:")
print(f"False Positives (non-acceptors predicted as acceptors): {fp}")
print(f"False Negatives (acceptors predicted as non-acceptors): {fn}")

if fn > fp:
    print("The model is more likely to miss actual loan acceptors (higher false negatives).")
elif fp > fn:
    print("The model is more likely to incorrectly predict acceptance for non-acceptors (higher false positives).")
else:
    print("Misclassification rates are balanced between the two classes.")
```

```python
#imports 
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# adding dataset 

ebay_df = pd.read_csv('../data/eBayAuctions.csv') # Krima
# ebay_df = pd.read_csv("eBayAuctions.csv") #Ihor
# auction_data = pd.read_csv('/Users/arturoperezespinosa/Downloads/eBayAuctions.csv') # Arthur

ebay_df.head()
```

3.a

```python
# Convert "Competitive?" to numeric 
ebay_df["Competitive"] = ebay_df["Competitive?"].apply(lambda x: 1 if x == 1 else 0)

print("HEAD:\n",ebay_df.head(), "\n" )
print("DESCRIPTION:\n", ebay_df.describe(include='all'))
```

```python
#pivot tables
pivot_category = ebay_df.pivot_table(index="Category", values="Competitive", aggfunc="mean")
pivot_currency = ebay_df.pivot_table(index="currency", values="Competitive", aggfunc="mean")
pivot_endday   = ebay_df.pivot_table(index="endDay", values="Competitive", aggfunc="mean")
pivot_duration = ebay_df.pivot_table(index="Duration", values="Competitive", aggfunc="mean")

print("PIVOT TABLES")
print("Mean of Competitive by Category:\n", pivot_category, "\n")
print("Mean of Competitive by currency:\n", pivot_currency, "\n")
print("Mean of Competitive by endDay:\n", pivot_endday, "\n")
print("Mean of Competitive by Duration:\n", pivot_duration,)    

# Generate pivot tables with for loop                                    ## we have to pick if we should use the top or the bottom one ##
# categorical_vars = ["Category", "currency", "endDay", "Duration"]
# for var in categorical_vars:
#     pivot_table = ebay_df.pivot_table(values="Competitive?", index=var, aggfunc="mean")
#     print(f"Pivot Table for {var}: {pivot_table}")

# Convert categorical features into dummy variables for logistic regression
ebay_df_dummies = pd.get_dummies( ebay_df, columns = ["Category", "currency", "endDay", "Duration"], drop_first=False)
```

```python
# Identify target and feature columns
# We'll assume "Competitive" is our target. If not, use "Competitive?" directly if it's 0/1.
if "Competitive" in ebay_df_dummies.columns:
    y = ebay_df_dummies["Competitive"]
else:
    y = ebay_df_dummies["Competitive?"]

# For the full model, we keep "ClosePrice" in the feature set
X_full = ebay_df_dummies.drop(columns=["Competitive", "Competitive?"], errors="ignore")
```

3.b

```python
X_train_full, X_val_full, y_train_full, y_val_full = train_test_split(
    X_full, y, test_size=0.4, random_state=123, stratify=y)

print("DATA SPLIT")
print("Training set size:", X_train_full.shape[0])
print("Validation set size:", X_val_full.shape[0])
```

```python
#Fit logistic regression with ALL predictors (including ClosePrice)
log_full = LogisticRegression(solver="liblinear", max_iter=1000)
log_full.fit(X_train_full, y_train_full)

# Evaluate on validation set
y_val_pred_full = log_full.predict(X_val_full)
acc_full = accuracy_score(y_val_full, y_val_pred_full)
cm_full = confusion_matrix(y_val_full, y_val_pred_full)

print("FULL MODEL")
print("Validation Accuracy (all predictors):", acc_full)
print("Confusion Matrix:\n", cm_full)
```

3.c

```python
X_no_price = X_full.drop(columns=["ClosePrice"])
X_train_np, X_val_np, y_train_np, y_val_np = train_test_split(
    X_no_price,
    y,
    test_size=0.4,
    random_state=123,
    stratify=y
)

log_no_price = LogisticRegression(solver="liblinear", max_iter=1000)
log_no_price.fit(X_train_np, y_train_np)

y_val_pred_np = log_no_price.predict(X_val_np)
acc_no_price = accuracy_score(y_val_np, y_val_pred_np)
cm_no_price = confusion_matrix(y_val_np, y_val_pred_np)

print("NO-PRICE MODEL")
print("Validation Accuracy (excluding ClosePrice):", acc_no_price)
print("Confusion Matrix:\n", cm_no_price)
```

```python
print("Comparison:")
print("  Full model accuracy:", acc_full)
print("  No-price accuracy:", acc_no_price)

# we can see the full model accuracy is better than the no-price accuracy
```

3.d

```python
# Identify the index of 'ClosePrice'
close_price_index = list(X_train_full.columns).index("ClosePrice")
close_price_coef  = log_full.coef_[0][close_price_index]
print("COEFFICIENT FOR CLOSEPRICE")
print("ClosePrice Coefficient:", close_price_coef, "\n")

# The closing Price (9.23 %) is statistically significant since the Coefficient
#   is less than significance level (10%).
```

3.e

```python
# Use L1-regularized logistic regression + GridSearchCV
log_l1 = LogisticRegression(penalty="l1", solver="liblinear", max_iter=1000)
param_grid = {"C": [0.01, 0.1, 1, 10, 100]}

grid_l1 = GridSearchCV(log_l1, param_grid, cv=5, scoring="accuracy")
grid_l1.fit(X_train_full, y_train_full)

best_l1_model = grid_l1.best_estimator_
y_val_pred_l1 = best_l1_model.predict(X_val_full)
acc_l1 = accuracy_score(y_val_full, y_val_pred_l1)

print("L1 Regularization")
print("Best C:", grid_l1.best_params_["C"])
print("Validation Accuracy with L1:", acc_l1)
print("Coefficients:\n", best_l1_model.coef_)
```
