```python
import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.preprocessing import MinMaxScaler

from sklearn.neural_network import MLPClassifier

from sklearn.metrics import (accuracy_score, confusion_matrix,
                             recall_score, precision_score,
                             f1_score, roc_auc_score,
                             classification_report)
```

```python
# data = pd.read_csv('../data/EastWestAirlinesNN.csv') # Artem

data = pd.read_csv('/Users/arturoperezespinosa/Downloads/EastWestAirlinesNN.csv') # Arthur

data.describe() 
```

```python
data.dropna(inplace=True)

data.dtypes
```

```python
X = data.drop(columns = ['Phone_sale'])

y = data['Phone_sale']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled  = scaler.transform(X_test)
```

```python
nnclassifier = MLPClassifier(hidden_layer_sizes=(5,),activation='logistic', max_iter=1000, early_stopping=True,  random_state=1)

nnclassifier.fit(X_train, y_train)
```

```python

pred_train = nnclassifier.predict(X_train_scaled) # Simplify predicted statement

pred_test = nnclassifier.predict(X_test_scaled)

print('Classification report for test data:')

print(classification_report(y_test, pred_test))

print('Classification report for train data:')

print(classification_report(y_train, pred_train))

print('Confusion matrix for test data:')

print(confusion_matrix(y_test, pred_test))
```

#### Part b

```python
nnclassifier_2 = MLPClassifier(hidden_layer_sizes=(1,),activation='logistic', max_iter=1000, early_stopping=True,  random_state=1)

nnclassifier_2.fit(X_train_scaled, y_train) # Using scaled training/test data instead
```

```python
pred_train_2 = nnclassifier_2.predict(X_train_scaled) # Simplified pred statement

pred_test_2 = nnclassifier_2.predict(X_test_scaled)

print('Classification report for test data:')

print(classification_report(y_test, pred_test_2)) # Second pred test

print('Classification report for train data:')

print(classification_report(y_train, pred_train_2)) # Second pred train

print('Confusion matrix for test data:')

print(confusion_matrix(y_test, pred_test_2))
```

Even with reduction in model complexity, it still struggles with the imbalance of the dataset! Hence, the change in hidden nodes doesn't affect the models performance noticeably!

#### Part c

```python
param_grid = {

    'hidden_layer_sizes': [(2,), (4,), (8,), (2, 2), (4, 4), (8, 8)],

    'activation': ['relu', 'logistic', 'tanh'],

    'batch_size': [16, 32, 64],

    'learning_rate_init': [0.01, 0.001, 0.0001],

    'max_iter': [100, 200, 300] # Went with higher max_iter because 20-80 wasn't enough for convergence

}

grid_search = GridSearchCV(MLPClassifier(random_state = 1, early_stopping = True), param_grid, cv = 3, n_jobs = -1, verbose = 2)

grid_search.fit(X_train_scaled, y_train)

print(f"Best parameters: {grid_search.best_params_}")
```

```python
best_model = grid_search.best_estimator_

pred_test = best_model.predict(X_test_scaled)   

print(f"Test Accuracy: {round(accuracy_score(y_test, pred_test), 3)}")

print(f"Test Recall: {round(recall_score(y_test, pred_test), 3)}")

print(f"Test Precision: {round(precision_score(y_test, pred_test), 3)}")

print(f"Test F1 Score: {round(f1_score(y_test, pred_test), 3)}")
```

If we apply class weights or resampling techniques, we will see significant improvement in the recall and precision for the minority class, which would improve F1 as well. This happens because we're mainly focusing on predicting the majority class.
