```python
# college2 = pd.read_csv('../data/College.csv')
Auto = pd.read_csv('../code1/data/Auto.csv') # Krima
```

```python
# removing missing data
Auto = pd.DataFrame(Auto)
print("Before removing missing data:", Auto.shape)
Auto2 = Auto.replace('?', np.nan).dropna()
print("After removing missing data:", Auto2.shape)
```

Question 1 (a)

```python
print("Before correcting predictor types:\n",Auto2.dtypes)

Auto2['horsepower'] = Auto2['horsepower'].astype(int)
print("\nNumber of unique values in origin column:", Auto2['origin'].nunique())
Auto2['origin'] = Auto2['origin'].astype(object)

print("\nAfter correcting predictor types:\n",Auto2.dtypes)
```

After removing all rows with missing data, we found that horsepower should be type int. Looking through the data we found that only 3 unique values in the origin column which means orgin is a numerical classification. The following shows which columns are quantitative and qualitative.

Quantitiative:
- mpg
- cylinders
- displacement

- year

Qualitative:
- origin

Question 2 (b)

```python
Quant_data = Auto2.drop(columns=['origin','name']) # Quantitiative data
# Qualitative = ['origin','name']
# Quantitiative = ['mpg', 'cylinders', 'displacement', 
#                  'horsepower', 'weight', 'acceleration', 
#                  'year']

for col in Quant_data.columns:
    col_min = np.min(Quant_data[col])  # Minimum value
    col_max = np.max(Quant_data[col])  # Maximum value
    col_range = col_max - col_min  # Range
    print(f"{col}: Min = {col_min}, Max = {col_max}, Range = {col_range}")
```

Question 3 (c)

```python
for col in Quant_data.columns:
    col_mean = Quant_data[col].mean()
    col_std = Quant_data[col].std()
    print(f"{col}: Mean = {col_mean:.3f}, Standard Deviation = {col_std:.3f}")
```

Question 4 (d)

```python
method1 = Quant_data.drop(index=Quant_data.iloc[10:85].index)
# method2 = Quant_data[(Quant_data.index < 10) | (Quant_data.index > 85)]

for col in method1.columns:
    col_min = np.min(method1[col])  # Minimum value
    col_max = np.max(method1[col])  # Maximum value
    col_range = col_max - col_min  # Range
    col_mean = method1[col].mean()
    col_std = method1[col].std()
    print(f"{col}: Range = {col_range:.2f}, Mean = {col_mean:.3f}, Standard Deviation = {col_std:.3f}")
```
