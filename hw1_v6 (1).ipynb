Q 2. a

```python
# load the dataset
data = pd.read_csv('/Users/arturoperezespinosa/Downloads/att.csv') # Arthur
print(data.dtypes)

# Picking variables for classifier training
att = data[["employment", "usage", "reachout", "card", "pick"] ]
print(att)
print(att.shape) # we see there are 1000 rows and 5 columns
```

Q 2.b

```python
missing_rate = att.isnull().mean() *100
missing_rate # the percent of null values in each attribute (column)
```

Q 2.c

```python
cleaned_att = att.dropna() # dropping all rows with missing values
print(cleaned_att)
print(cleaned_att.shape) # we see there are 955 rows and 5 columns
```

Q 2.d

```python
# Stratifying test and train data

train_set, test_set = train_test_split(cleaned_att, test_size = 0.2, random_state = 100, stratify = cleaned_att["pick"]) # Using 80% of the data for training purposes
```

Q 2.e

```python
# Encoding qualitative variables to numeric variables 

categorical_features = ["employment", "usage", "reachout", "card"]

ordinal_encoder = OrdinalEncoder(handle_unknown = "use_encoded_value", unknown_value = -1) # Handle unmapped values to continue processing

train_set.loc[:, categorical_features] = ordinal_encoder.fit_transform(train_set[categorical_features])

test_set.loc[:, categorical_features] = ordinal_encoder.transform(test_set[categorical_features])
```

```python
# Encoding class variable to numeric variable 

label_encoder = LabelEncoder()

train_set.loc[:, "pick"] = label_encoder.fit_transform(train_set["pick"])

test_set.loc[:, "pick"] = label_encoder.transform(test_set["pick"])
```

Q 2.f

```python
# 7NN classifier training

X_train, y_train = train_set.drop(columns = ["pick"]), train_set["pick"] # Model learning

X_test, y_test = test_set.drop(columns=["pick"]), test_set["pick"] # Model evaluation

y_train = y_train.astype(int) # Convert to integer for training

y_test = y_test.astype(int) # Convert to integer for training

knn = KNN(n_neighbors = 7)

knn.fit(X_train, y_train)

#  Model accuracy evaluation

accuracy = knn.score(X_test, y_test) # Quantify model accuracy with score method

print(f"7NN Model Accuracy: {accuracy:.3f}")
```
